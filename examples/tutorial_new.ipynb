# Intro to pyNS

pyNS is a Python library to programmatically access the Neuroscout API. pyNS let's you query the API and create analyses, without having to mess around with buildling any JSON requests yourself.

In this tutorial, I'll demonstrate how to query the API to create your own analysis.

```python
from pyns import Neuroscout
api = Neuroscout()
```

```python
datasets=api.datasets.get()
```

```python
#how many datasets are there?
print(f'dataset count = {len(api.datasets.get())}\n')
```

```python
#list the datasets and IDs?
print('Datasets and IDs:\n')
for i in datasets:
    print(i['name'], i['id'])
```

```python
dataset = datasets[1]
```

```python
#what are the keys of the dataset dictionary?
dataset.keys()
```

```python
api.runs.get(dataset_id=dataset['id']) #get the run ids. run ids are important because some datasets will have different stimuli or broke their stimuli up into multiple parts
```
